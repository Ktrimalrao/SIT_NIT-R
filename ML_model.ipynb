```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
X = data.data
y = data.target
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score
```

```python
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
DT = DecisionTreeClassifier()
DT.fit(X_train,y_train)
```

```python
RF = RandomForestClassifier()
RF.fit(X_train,y_train)
```

```python
NB = GaussianNB()
NB.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
acc1 = accuracy_score(y_test,y_pred1)
print("Logistic Regression Accuracy : ",acc1)
```

```python
y_pred2 = DT.predict(X_test)
acc2 = accuracy_score(y_test,y_pred2)
print("Decision Tree Classifier Accuracy : ",acc2)
```

```python
y_pred3 = RF.predict(X_test)
acc3 = accuracy_score(y_test,y_pred3)
print("Random Forest Classifier Accuracy : ",acc3)
```

```python
y_pred4 = NB.predict(X_test)
acc4 = accuracy_score(y_test,y_pred4)
print("Naive Bayes Accuracy : ",acc4)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test,y_pred3)
mat
```

```python
sns.heatmap(data=mat,annot=True)
plt.show()
```
